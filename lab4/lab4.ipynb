```python
def barCoords(triangle, x, y):
    x0 = triangle[0][0]
    x1 = triangle[1][0]
    x2 = triangle[2][0]
    y0 = triangle[0][1]
    y1 = triangle[1][1]
    y2 = triangle[2][1]
    lambda0 = ((x1 - x2)*(y - y2) - (y1 - y2)*(x - x2)) / \
        ((x1 - x2)*(y0 - y2) - (y1 - y2)*(x0 - x2))
    lambda1 = ((x2 - x0)*(y - y0) - (y2 - y0)*(x - x0)) / \
        ((x2 - x0)*(y1 - y0) - (y2 - y0)*(x1 - x0))
    lambda2 = ((x0 - x1)*(y - y1) - (y0 - y1)*(x - x1)) / \
        ((x0 - x1)*(y2 - y1) - (y0 - y1)*(x2 - x1))
    return (lambda0 >= 0. and lambda1 >= 0. and lambda2 >= 0.)
```

```python
def boundRect(triangle, w, h):
    xleft = min(triangle[0][0], triangle[1][0], triangle[2][0]) - 1 
    xright = max(triangle[0][0], triangle[1][0], triangle[2][0]) + 1
    ytop = min(triangle[0][1], triangle[1][1], triangle[2][1]) - 1
    ybot = max(triangle[0][1], triangle[1][1], triangle[2][1]) + 1
    if (xleft < 0):
        xleft = 0
    if (ytop < 0):
        ytop = 0
    if (xright > w):
        xright = w
    if (ybot > h):
        ybot = h
    lst = [xleft, xright, ytop, ybot]
    return [int(i) for i in lst]
```

```python
def normal(triangle):
    a = [triangle[1][0] - triangle[0][0],
         triangle[1][1] - triangle[0][1],
         triangle[1][2] - triangle[0][2]]
    b = [triangle[1][0] - triangle[2][0],
         triangle[1][1] - triangle[2][1],
         triangle[1][2] - triangle[2][2]]
    return np.cross(a, b)
```

```python
normals = np.ndarray((len(triangles), 3), dtype=float)
for i in range(len(triangles)):
    normals[i] = normal(triangles[i])
```

```python
def barCoords1(triangle, x, y):
    x0 = triangle[0][0]
    x1 = triangle[1][0]
    x2 = triangle[2][0]
    y0 = triangle[0][1]
    y1 = triangle[1][1]
    y2 = triangle[2][1]
    lambda0 = ((x1 - x2)*(y - y2) - (y1 - y2)*(x - x2)) / \
        ((x1 - x2)*(y0 - y2) - (y1 - y2)*(x0 - x2))
    lambda1 = ((x2 - x0)*(y - y0) - (y2 - y0)*(x - x0)) / \
        ((x2 - x0)*(y1 - y0) - (y2 - y0)*(x1 - x0))
    lambda2 = ((x0 - x1)*(y - y1) - (y0 - y1)*(x - x1)) / \
        ((x0 - x1)*(y2 - y1) - (y0 - y1)*(x2 - x1))
    return (lambda0 >= 0. and lambda1 >= 0. and lambda2 >= 0.), lambda0, lambda1, lambda2
```

```python
def projTrans(triangle, ax, ay, u0, v0, tz):
    A = np.array([[ax, 0, u0], [0, -ay, v0], [0, 0, 1]],  dtype=float)
    coords = np.array(triangle)
    coords[:, 2] +=tz
    res = np.zeros((len(coords), 3), dtype=float)
    for i in range(len(coords)):
        res[i] = np.dot(A, coords[i])
        res[i][:2] /= res[i][2]    
    return res
```

```python
z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(cs)
for i in range(len(triangles)):
    if(cos[i] < 0):
        r = -255 * cos[i]
        g = -255 * cos[i]
        b = -255 * cos[i]
        newTr = projTrans(triangles[i], 1500, 1500, 500, 500, 300)
        xleft, xright, ytop, ybot = boundRect(newTr, w, h)
        for x in range(xleft, xright):
            for y in range(ytop, ybot):
                checkZero, lambda0, lambda1, lambda2 = barCoords1(
                    newTr, x, y)
                if(checkZero):
                    _z = lambda0*newTr[0][2]+lambda1 * \
                        newTr[1][2]+lambda2*newTr[2][2]
                    if(_z < z[x, y]):
                        image[x, y] = [r, g, b]
                        z[x, y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image6.png")
```

```python
L = [1, 0, 1]

z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(cs)
alpha = 5*math.pi/4
beta = -math.pi/6
gamma = -math.pi/2
RY = np.array([[math.cos(beta), 0, math.sin(beta)], [
    0, 1, 0], [-math.sin(beta), 0, math.cos(beta)]])
RX = np.array([[1, 0, 0], [0, math.cos(alpha), math.sin(alpha)],
               [0, -math.sin(alpha), math.cos(alpha)]])
RZ = np.array([[math.cos(gamma), math.sin(gamma), 0], [
    -math.sin(gamma), math.cos(gamma), 0], [0, 0, 1]])

rot = []
for i in range(len(triangles)):
    rot.append(RZ.dot(np.transpose(triangles[i])))

newTrs = []
for i in range(len(rot)):
    rotTr = RX.dot(rot[i])
    newTrs.append(projTrans(np.transpose(rotTr), 1500, 1500, 500, 500, 300))    
```

```python
newNormals = np.ndarray((len(newTrs), 3), dtype=float)
for i in range(len(newTrs)):
    newNormals[i] = normal(newTrs[i])
```

```python
pointsSet = set()
for i in range(len(newTrs)):
    for j in range(3):
        pointsSet.add(str(newTrs[i][j]))
```

```python
normalsPoints = {x: [0, 0, 0] for x in pointsSet}
for i in range(len(newTrs)):
    for j in range(3):
        normalsPoints[str(newTrs[i][j])] += newNormals[i]
for i in normalsPoints:
    normalsPoints[i] /= np.linalg.norm(normalsPoints[i])
```

```python
for i in range(len(newTrs)):
    Li = []
    for k in range(3):
        Li.append(np.dot(normalsPoints[str(newTrs[i][k])], L) /
                  (np.linalg.norm(
                      normalsPoints[str(newTrs[i][k])])*np.linalg.norm(L)))
    xleft, xright, ytop, ybot = boundRect(newTrs[i], w, h)
    for x in range(xleft, xright):
        for y in range(ytop, ybot):
            checkZero, lambda0, lambda1, lambda2 = barCoords1(
                newTrs[i], x, y)
            if(checkZero):
                _z = lambda0*newTrs[i][0][2]+lambda1 * \
                    newTrs[i][1][2]+lambda2*newTrs[i][2][2]
                if(_z < z[x, y]):
                    a = -255 * (lambda0*Li[0] + lambda1*Li[1] + lambda2*Li[2])
                    if (a > 0):
                        image[x, y] = a
                        z[x, y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image8.png")
```
