# 8. Барицентрические координаты
Написать функцию вычисления барицентрических координат для точки с экранными (целочисленными координатами) (x,y) относительно вещественных вершин треугольника (x0, y0), (x1, y1) и (x2, y2).
Они вычисляются по формулам:

Убедиться, что сумма барицентрических координат
lambda0 + lambda1 + lambda2 = 1.0
Обратите внимание, что координаты (x, y) – экранные, и поэтому целочисленные. В то же время вершины треугольника (x0, y0), (x1, y1) и (x2, y2) – вещественные, округлять их перед вычислениями не надо.
С точки зрения написания и выполнения программы это не оказывает большого влияния, но важно для понимания. Барицентрические координаты в рамках этого курса будут описывать положение пикселя, который вы собираетесь отрисовать, относительно реального треугольника.

```python
def barCoords(triangle, x, y):
    x0 = triangle[0][0]
    x1 = triangle[1][0]
    x2 = triangle[2][0]
    y0 = triangle[0][1]
    y1 = triangle[1][1]
    y2 = triangle[2][1]
    lambda0 = ((x1 - x2)*(y - y2) - (y1 - y2)*(x - x2)) / \
        ((x1 - x2)*(y0 - y2) - (y1 - y2)*(x0 - x2))
    lambda1 = ((x2 - x0)*(y - y0) - (y2 - y0)*(x - x0)) / \
        ((x2 - x0)*(y1 - y0) - (y2 - y0)*(x1 - x0))
    lambda2 = ((x0 - x1)*(y - y1) - (y0 - y1)*(x - x1)) / \
        ((x0 - x1)*(y2 - y1) - (y0 - y1)*(x2 - x1))
    return (lambda0 >= 0. and lambda1 >= 0. and lambda2 >= 0.)
```

Разумеется, вы можете сделать 1 и 2 пункты одновременно.
3. Для каждого пикселя внутри ограничивающего прямоугольника вычислить барицентрические координаты относительно вершин треугольника.
Если все барицентрические координаты пикселя больше нуля – пиксель рисуется, иначе – переходим к следующему.
Обратите внимание, что рёбра треугольника линиями (как в задании 7) рисовать не надо.

```python
def boundRect(triangle, w, h):
    xleft = min(triangle[0][0], triangle[1][0], triangle[2][0]) - 1 
    xright = max(triangle[0][0], triangle[1][0], triangle[2][0]) + 1
    ytop = min(triangle[0][1], triangle[1][1], triangle[2][1]) - 1
    ybot = max(triangle[0][1], triangle[1][1], triangle[2][1]) + 1
    if (xleft < 0):
        xleft = 0
    if (ytop < 0):
        ytop = 0
    if (xright > w):
        xright = w
    if (ybot > h):
        ybot = h
    lst = [xleft, xright, ytop, ybot]
    return [int(i) for i in lst]
```

# 12. Вычисление нормали к поверхности треугольника
Для каждого треугольника вычислить нормаль к этому треугольнику по формуле:

где (𝑋0,𝑌0,𝑍0), (𝑋1,𝑌1,𝑍1) и (𝑋2,𝑌2,𝑍2) – исходные координаты вершин треугольника (до любых преобразований), а × – векторное произведение.
Координаты нормали могут быть вычислены через определитель:

    

```python
def normal(triangle):
    a = [triangle[1][0] - triangle[0][0],
         triangle[1][1] - triangle[0][1],
         triangle[1][2] - triangle[0][2]]
    b = [triangle[1][0] - triangle[2][0],
         triangle[1][1] - triangle[2][1],
         triangle[1][2] - triangle[2][2]]
    return np.cross(a, b)
```

```python
normals = np.ndarray((len(triangles), 3), dtype=float)
for i in range(len(triangles)):
    normals[i] = normal(triangles[i])
```

Для каждого треугольника определить косинус угла падения направленного света (считать направление света равным 𝑙̅=[0,0,1]) через нормализованное скалярное произведение 〈𝑛̅,𝑙̅〉‖𝑛̅‖∙‖𝑙̅‖.
Изменить цикл отрисовки полигонов таким образом, чтобы отрисовывались только полигоны с 〈𝑛̅,𝑙̅〉‖𝑛̅‖∙‖𝑙̅‖<0.

```python
L = [0, 0, 1]
cos = np.zeros(len(normals), dtype=float)
for i in range(len(normals)):
    cos[i] = np.dot(normals[i], L) / \
        (np.linalg.norm(normals[i])*np.linalg.norm(L))
```

# 15. z-буфер
При отрисовке полигонов проверять перекрытие полигонов с использованием z-буфера.
z-буфер – это матрица из вещественных значений по размеру совпадающая с изображением. Все элементы z-буфера изначально инициализируются некоторым достаточно большим значением.
При отрисовке для каждой точки выполняется следующая проверка:
1. Вычисляются барицентрические координаты (𝜆0, 𝜆1, 𝜆2).
2. Если все барицентрические координаты больше нуля, вычисляем z-координату исходного полигона через исходные z-координаты вершин этого полигона:
    𝑧̂= 𝜆0𝑧0+𝜆1𝑧1+𝜆2𝑧2.
3. Если вычисленное значение координаты 𝑧̂ больше координаты z-буфера для текущего пикселя, пропускаем точку.
4. Если вычисленное значение координаты 𝑧̂ меньше координаты z-буфера для текущего пикселя, отрисовываем этот пиксель, а соответствующему элементу z-буфера присваиваем значение 𝑧̂.

```python
def barCoords1(triangle, x, y):
    x0 = triangle[0][0]
    x1 = triangle[1][0]
    x2 = triangle[2][0]
    y0 = triangle[0][1]
    y1 = triangle[1][1]
    y2 = triangle[2][1]
    lambda0 = ((x1 - x2)*(y - y2) - (y1 - y2)*(x - x2)) / \
        ((x1 - x2)*(y0 - y2) - (y1 - y2)*(x0 - x2))
    lambda1 = ((x2 - x0)*(y - y0) - (y2 - y0)*(x - x0)) / \
        ((x2 - x0)*(y1 - y0) - (y2 - y0)*(x1 - x0))
    lambda2 = ((x0 - x1)*(y - y1) - (y0 - y1)*(x - x1)) / \
        ((x0 - x1)*(y2 - y1) - (y0 - y1)*(x2 - x1))
    return (lambda0 >= 0. and lambda1 >= 0. and lambda2 >= 0.), lambda0, lambda1, lambda2
```

```python
z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(coords)

for i in range(len(triangles)):
    if(cos[i] < 0):
        r = -255 * cos[i]
        g = -255 * cos[i]
        b = -255 * cos[i]
        xleft, xright, ytop, ybot = boundRect(triangles[i], w, h)
        for x in range(xleft, xright):
            for y in range(ytop, ybot):
                checkZero, lambda0, lambda1, lambda2 = barCoords1(
                    triangles[i], x, y)
                if(checkZero):
                    _z = lambda0*triangles[i][0][2]+lambda1 * \
                        triangles[i][1][2]+lambda2*triangles[i][2][2]
                    if(_z < z[x][y]):
                        image[x, y] = [r, g, b]
                        z[x][y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image5.png")
```

# 16. Проективное преобразование
Заменить преобразование к экранным координатам
[4000 * X + 500, 4000 * Y + 500]
на проективное преобразование: 

где (𝑢, 𝑣) – экранные координаты, 𝑎𝑥, 𝑎𝑦 – масштаб, (𝑢0, 𝑣0) – центр изображения.
Чтобы проективное преобразование к координате Z всех точек модели необходимо добавить фиксированное значение (не менее 0.1).


```python
def projTrans(triangle, ax, ay, u0, v0, tz):
    A = np.array([[ax, 0, u0], [0, -ay, v0], [0, 0, 1]],  dtype=float)
    coords = np.array(triangle)
    coords[:, 2] +=tz
    res = np.zeros((len(coords), 3), dtype=float)
    for i in range(len(coords)):
        res[i] = np.dot(A, coords[i])
        res[i][:2] /= res[i][2]    
    return res
```

```python
z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(cs)
for i in range(len(triangles)):
    if(cos[i] < 0):
        r = -255 * cos[i]
        g = -255 * cos[i]
        b = -255 * cos[i]
        newTr = projTrans(triangles[i], 1500, 1500, 500, 500, 300)
        xleft, xright, ytop, ybot = boundRect(newTr, w, h)
        for x in range(xleft, xright):
            for y in range(ytop, ybot):
                checkZero, lambda0, lambda1, lambda2 = barCoords1(
                    newTr, x, y)
                if(checkZero):
                    _z = lambda0*newTr[0][2]+lambda1 * \
                        newTr[1][2]+lambda2*newTr[2][2]
                    if(_z < z[x, y]):
                        image[x, y] = [r, g, b]
                        z[x, y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image6.png")
```

# 17. Поворот модели:
Реализовать поворот модели. Для этого исходные координаты вершин заменить на преобразованные:

Для начала рекомендуется попробовать реализовать поворот вокруг оси Y. Такой поворот проще всего визуально отследить и проверить.

```python
L = [1, 0, 1]
cos = np.zeros(len(normals), dtype=float)
for i in range(len(normals)):
    cos[i] = np.dot(normals[i], L) / \
        (np.linalg.norm(normals[i])*np.linalg.norm(L))

z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(cs)
alpha = 5*math.pi/4
beta = -math.pi/6
gamma = -math.pi/2
RY = np.array([[math.cos(beta), 0, math.sin(beta)], [
    0, 1, 0], [-math.sin(beta), 0, math.cos(beta)]])
RX = np.array([[1, 0, 0], [0, math.cos(alpha), math.sin(alpha)],
               [0, -math.sin(alpha), math.cos(alpha)]])
RZ = np.array([[math.cos(gamma), math.sin(gamma), 0], [
    -math.sin(gamma), math.cos(gamma), 0], [0, 0, 1]])

rot = []
for i in range(len(triangles)):
    rot.append(RZ.dot(np.transpose(triangles[i])))

for i in range(len(rot)):
    if(cos[i] < 0):
        r = -255 * cos[i]
        g = -255 * cos[i]
        b = -255 * cos[i]

        rotTr = RX.dot(rot[i])
        newTr = projTrans(np.transpose(rotTr), 1500, 1500, 500, 500, 300)
        xleft, xright, ytop, ybot = boundRect(newTr, w, h)
        for x in range(xleft, xright):
            for y in range(ytop, ybot):
                checkZero, lambda0, lambda1, lambda2 = barCoords1(
                    newTr, x, y)
                if(checkZero):
                    _z = lambda0*newTr[0][2]+lambda1 * \
                        newTr[1][2]+lambda2*newTr[2][2]
                    if(_z < z[x, y]):
                        image[x, y] = [r, g, b]
                        z[x, y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image7.png")
```

# 18. Затенение Гуро.
Реализовать продвинутое освещение посредством метода затенения Гуро.

Для этого необходимо будет использовать нормали ко всем вершинам модели (можно загрузить из модели либо вычислить, усреднив нормали ко всем полигонам, соседним с вершиной).

Для каждой вершины вычислить освещение этой вершины (как в задании 14).

Для каждой точки полигона вычислять освещение через барицентрические координаты относительно вершин этого полигона:
Пусть 𝑙0,𝑙1,𝑙2 – соответственно равны


```python
L = [1, 0, 1]

z = np.full((h, w), np.inf)
image = np.zeros((h, w, 3), dtype=np.uint8)
triangles = toTriangles(cs)
alpha = 5*math.pi/4
beta = -math.pi/6
gamma = -math.pi/2
RY = np.array([[math.cos(beta), 0, math.sin(beta)], [
    0, 1, 0], [-math.sin(beta), 0, math.cos(beta)]])
RX = np.array([[1, 0, 0], [0, math.cos(alpha), math.sin(alpha)],
               [0, -math.sin(alpha), math.cos(alpha)]])
RZ = np.array([[math.cos(gamma), math.sin(gamma), 0], [
    -math.sin(gamma), math.cos(gamma), 0], [0, 0, 1]])

rot = []
for i in range(len(triangles)):
    rot.append(RZ.dot(np.transpose(triangles[i])))

newTrs = []
for i in range(len(rot)):
    rotTr = RX.dot(rot[i])
    newTrs.append(projTrans(np.transpose(rotTr), 1500, 1500, 500, 500, 300))    
```

```python
newNormals = np.ndarray((len(newTrs), 3), dtype=float)
for i in range(len(newTrs)):
    newNormals[i] = normal(newTrs[i])
```

```python
pointsSet = set()
for i in range(len(newTrs)):
    for j in range(3):
        pointsSet.add(str(newTrs[i][j]))
```

```python
normalsPoints = {x: [0, 0, 0] for x in pointsSet}
for i in range(len(newTrs)):
    for j in range(3):
        normalsPoints[str(newTrs[i][j])] += newNormals[i]
for i in normalsPoints:
    normalsPoints[i] /= np.linalg.norm(normalsPoints[i])
```

```python
for i in range(len(newTrs)):
    Li = []
    for k in range(3):
        Li.append(np.dot(normalsPoints[str(newTrs[i][k])], L) /
                  (np.linalg.norm(
                      normalsPoints[str(newTrs[i][k])])*np.linalg.norm(L)))
    xleft, xright, ytop, ybot = boundRect(newTrs[i], w, h)
    for x in range(xleft, xright):
        for y in range(ytop, ybot):
            checkZero, lambda0, lambda1, lambda2 = barCoords1(
                newTrs[i], x, y)
            if(checkZero):
                _z = lambda0*newTrs[i][0][2]+lambda1 * \
                    newTrs[i][1][2]+lambda2*newTrs[i][2][2]
                if(_z < z[x, y]):
                    a = -255 * (lambda0*Li[0] + lambda1*Li[1] + lambda2*Li[2])
                    if (a > 0):
                        image[x, y] = a
                        z[x, y] = _z

img = PIL.Image.fromarray(image, 'RGB')
img.save("image8.png")
```
